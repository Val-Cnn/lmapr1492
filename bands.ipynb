# 1. Détermination de la bande interdite
La bande interdite est l'écart d'énergie se trouvant entre la bande de valence et la bande de conduction, elle constitue un ensemble de niveaux d'énergies qui ne peuvent être occupés par des électrons. 
Lorsque un cristal est métallique, les électrons sont dits "libres" car ils peuvent se déplacer librement dans l'ensemble du cristal sans être soumis à un quelconque potentiel. Dans le cas de mon cristal de GaSb, il fait partie de cette classe, il n'y a donc pas de bande interdite qu'elle soit directe ou indirecte.

# 3. Masse effective
Il est souvent intéressant de décrire les électrons comme des
particules libres dont la masse effective varie (plutôt que des
particules de masse fixée évoluant dans un potentiel).
Cette masse effective peut-être positive ou négative, supérieure
ou inférieure à la masse réelle de l’électron.
En toute généralité, la masse effective est définie par le tenseur
d'ordre 2 des dérivées secondes de l’énergie E par rapport au
vecteur d’onde $\bf{k}$ :
$m_{ij}^* = (\frac{\delta E^2}{\delta k_i \delta k_j})^{-1}$


Dans le cas des électrons de ce métal, il n'y a ni bande de valence ni bande de conduction comme elles sont toutes deux confoncdues, les électrons étant libres, leur masse effective est simplement la masse d'un électron dans le vide, soit $9,109*10^{−31} kg$
