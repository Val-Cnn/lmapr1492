# Diffractogramme du cristal de GaSb

Nous avons un cristal de GaSb, nous allons déterminer les 3 premiers pics ainsi que leur direction correspondante

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.analysis.bond_valence import BVAnalyzer
import numpy as np
```

```python
with MPRester("SmhzvzkiqMXecCh0o4") as m:

    # Structure de mon cristal, GaSb
    struct = m.get_structure_by_material_id("mp-1018059")
```

# Rappel

Notons la position des atomes dans la maille de GaSb ${\bf r}_j$, l'intensité des pics est proportionnelle au facteur de structure de la maille, soit, $S_k = \sum\limits_{j=1}^4  f_j ({\bf K}) e^{i{\bf K}\cdot{\bf r}_j'}$ où $f_j ({\bf K})$ est le facteur de forme atomique de l'atome.
 ${\bf K}$ est un vecteur du réseau réciproque défni comme $ {\bf K} = h{\bf a}^*+k{\bf b}^*+l{\bf c}^*$ avec ${\bf a}^*, {\bf b}^*, {\bf c}^*$ les vecteurs du réseau réciproque et $h, k, l$ les indices qui définissent le vecteur d'onde.
$S_k = f_{Ga}*(e^{hx_1+ky_1+lz_1}+e^{hx_3+ky_3+lz_3})+f_{Sb}*(e^{hx_2+ky_2+lz_2}+e^{hx_4+ky_4+lz_4})$
$S_k = f_{Ga}*(e^{2\pi(2h/3+k/3+z/2)}+e^{2\pi(2h/3+k/3+7z/8)})+f_{Sb}*(e^{2\pi(h/3+2k/3)}+e^{2\pi(h/3+2k/3+3l/8}))$

$f_{Ga}$ et $f_{Sb}$ sont les facteurs de formes des deux types d'atomes constituants la maille.

L'intensité des rayons diffusés pour un pic est donnée par $|S_{\bf{k}}|^2$. 

Au niveau de l'angle $2\theta$, il est déduit par la condition de Bragg, $n\lambda=2d_{hkl}\sin(\theta)$ où $\lambda$ est l longueur d'onde du rayon de diffraction et $d$ vaut la distance entre deux plans $(hkl)$ successifs, soit $d_{hkl}=2\pi/|\bf{K}|$, avec ${\bf{K}}=(h,k,l)=h{\bf{a^*}}+k{\bf{b^*}}+l{\bf{c^*}}$



```python
#longueur d'onde du rayon X
lambd = 1.5406 
#Fonction qui créé un calculateur de diffractogramme
xrd = XRDCalculator(lambd) 
# Calcule le diffractogramme du GaSb
Xcalc= xrd.get_pattern(structure=struct, scaled=False, two_theta_range=(0,30))
XcalcSCALED= xrd.get_pattern(structure=struct, scaled=True, two_theta_range=(0,30))
# Affiche le diffractogramme du GaSb
xrd.get_plot(struct, two_theta_range=(0,30))
```

## Premier pic
On remarque que la multiplicité de ce pic est de 6 et l'angle theta est le plus petit des 3, celà joue certainement dans son intensité qui est la plus grande des 3.

```python
print("L'angle prédit pour le plan de diffraction ci dessus est de " + str(Xcalc.x[0]))
print("L'intensité des rayons diffusés pour ce premier pic est de "+ str(Xcalc.y[0]))
print("Une fois mit à l'échelle par rapport aux deux autres, on obtient pour le premier pic : " + str(XcalcSCALED.y[0]))
print("Les indices associés à ce premier pic sont" + str(Xcalc.hkls[0]))
print("La distance entre deux plans portant ces indices est " + str(Xcalc.d_hkls[0]))
```

## Deuxième pic
On remarque que la multiplicité de ce pic est de 2, ce qui est joue certainement, en plus de l'angle theta plus grand, dans sa plus faible intensité par rapport au premier.

```python
print("L'angle prédit pour le plan de diffraction ci dessus est de " + str(Xcalc.x[1]))
print("L'intensité des rayons diffusés pour ce premier pic est de "+ str(Xcalc.y[1]))

print("Une fois mit à l'échelle par rapport aux deux autres, on obtient pour le premier pic : " + str(XcalcSCALED.y[1]))
print("Les indices associés à ce premier pic sont" + str(Xcalc.hkls[1]))
print("La distance entre deux plans portant ces indices est " + str(Xcalc.d_hkls[1]))
```

## Troisième pic
On remarque que la multiplicité de ce pic est de 12 et l'angle theta est le plus grand des 3. Le fait que son intensité diffusée soit supérieure à celle du deuxième pic est du à sa grande multiplicité qui renforce le rayon diffracté.

```python
print("L'angle prédit pour le plan de diffraction ci dessus est de " + str(Xcalc.x[2]))
print("L'intensité des rayons diffusés pour ce premier pic est de "+ str(Xcalc.y[2]))

print("Une fois mit à l'échelle par rapport aux deux autres, on obtient pour le premier pic : " + str(XcalcSCALED.y[2]))
print("Les indices associés à ce premier pic sont" + str(Xcalc.hkls[2]))
print("La distance entre deux plans portant ces indices est " + str(Xcalc.d_hkls[2]))
```
